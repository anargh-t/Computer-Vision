```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([
    [0, 0],
    [1, 0],

])

def translate(points, tx, ty):
    return points + np.array([tx, ty])

def rotate(points, angle):
    rotation_matrix = np.array([
        [np.cos(angle), -np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])
    return np.dot(points, rotation_matrix.T)

def scale(points, sx, sy):
    scaling_matrix = np.array([
        [sx, 0],
        [0, sy]
    ])
    return np.dot(points, scaling_matrix.T)

points_translated = translate(points, 2, 1)

points_rotated = rotate(points, np.pi/4)

points_scaled = scale(points, .5, .5)

print("Original Points:")
print(points)
print("\nTranslated Points:")
print(points_translated)
print("\nRotated Points:")
print(points_rotated)
print("\nScaled Points:")
print(points_scaled)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_translated[:, 0], points_translated[:, 1], 'o-', label='Translated')
plt.title('Translation')
plt.legend()
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_rotated[:, 0], points_rotated[:, 1], 'o-', label='Rotated')
plt.title('Rotation')
plt.legend()
plt.axis('equal')

plt.subplot(1, 3, 3)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_scaled[:, 0], points_scaled[:, 1], 'o-', label='Scaled')
plt.title('Scaling')
plt.legend()
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
shear_matrix = np.array([
    [1, 0.5],
    [0, 1]
])
points_sheared = np.dot(points, shear_matrix.T)
points_affine = translate(points_sheared, 1, 0.5)

points_similarity = scale(rotate(points, np.pi/6), 0.75, 0.75)
points_similarity = translate(points_similarity, -1, -1)

points_euclidean = rotate(points, -np.pi/4)
points_euclidean = translate(points_euclidean, -2, 2)

print("Original Points:")
print(points)
print("\nAffine Points:")
print(points_affine)
print("\nSimilarity Points:")
print(points_similarity)
print("\nEuclidean Points:")
print(points_euclidean)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_affine[:, 0], points_affine[:, 1], 'o-', label='Affine')
plt.title('Affine Transformation')
plt.legend()
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_similarity[:, 0], points_similarity[:, 1], 'o-', label='Similarity')
plt.title('Similarity Transformation')
plt.legend()
plt.axis('equal')

plt.subplot(1, 3, 3)
plt.plot(points[:, 0], points[:, 1], 'o-', label='Original')
plt.plot(points_euclidean[:, 0], points_euclidean[:, 1], 'o-', label='Euclidean')
plt.title('Euclidean Transformation')
plt.legend()
plt.axis('equal')

plt.tight_layout()
plt.show()
```
