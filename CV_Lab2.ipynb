# Convert world coordinates to pixel coordinates

1. Define camera intrinsics
2. camera extrnsics
3. Define world coordinates
4. Transform world point to camera coordinates
5. Project to image plane
6. Convert homogenous coordinates to pixel coordinates

```python
# Camera Intrinsics Matrix (K)
K = np.array([[880, 0, 320],
              [0, 800, 240],
              [0, 0, 1]])

# Rotation Matrix (R) - World to Camera
R = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])

# Translation Vector (T) - World to Camera
T = np.array([0, 0, 5]).reshape(3, 1)

# World Coordinates
World = np.array([1, 2, 3])
```

```python
# Convert world coordinates to homogeneous coordinates
World_homogeneous = np.append(World, 1).reshape(4, 1)

# Create [R|T] matrix
RT = np.hstack((R, T))  # shape: 3x4

# Project world coordinates to camera coordinates
Camera_coords = RT @ World_homogeneous  # shape: 3x1

# Project camera coordinates to homogeneous pixel coordinates
Pixel_coords_homogeneous = K @ Camera_coords  # shape: 3x1

# Convert homogeneous to pixel coordinates
Pixel_coords = Pixel_coords_homogeneous[:2] / Pixel_coords_homogeneous[2]

print(Pixel_coords.flatten())
```
